# Pre-processing

```python
import pandas as pd

df = pd.read_excel("Netflix Dataset Latest 2021.xlsx")
columns = ['Title', 'Genre', 'Tags', 'Director', 'Writer', 'Actors']
data = df[columns]
if data.isnull().values.any(): data=data.dropna().reset_index(drop=True)
features = [(str(data['Title'][i]) + ' ' + str(data['Genre'][i]) + ' ' + str(data['Tags'][i]) + ' ' + 
             str(data['Director'][i]) + ' ' + str(data['Writer'][i]) + ' ' + str(data['Actors'][i]))
            for i in range(data.shape[0])]
data['Features'] = features
data['Movie_id'] = [i for i in range(data.shape[0])]
```

```python
data
```

# Training

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vectorizer = CountVectorizer()
#vectorizer = TfidfVectorizer()
cm = vectorizer.fit_transform(data['Features'])
vectorizer.get_feature_names_out()
```

```python
cs = cosine_similarity(cm)
cs.shape
```

# Test

```python
title = 'Snowroller'
movie_id = data[data.Title == title]['Movie_id'].values[0]
movie_id
```

```python
scores = list(enumerate(cs[movie_id])) # list(movie_id, similarity score)
scores
```

```python
sorted_scores = sorted(scores, key=lambda x:x[1], reverse=True)
sorted_scores = sorted_scores[1:]
sorted_scores
```

```python
print(f"The 7 most recommened movies to {title} are:")
idx = 0
movie_list = pd.unique(data['Movie_id'])

for score in sorted_scores:
    if idx == 7: break
    if score[0] in movie_list:
        idx += 1
        print(idx, data[data.Movie_id == score[0]]['Title'].values[0])
```
